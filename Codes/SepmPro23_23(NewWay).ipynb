```python
import re
```

```python
descripton_column = ['TITLE', 'BULLET_POINTS', 'DESCRIPTION']
html_pattern = re.compile('<.*?>')
```

```python
for column in descripton_column:
    # Check if the column has any non-null values
    if train_data[column].notnull().any():
        # Remove special symbols
        train_data[column] = train_data[column].apply(lambda x: re.sub('[^a-zA-Z0-9\s]', '', x) if pd.notnull(x) else x)

        # Remove HTML tags
        train_data[column] = train_data[column].apply(lambda x: re.sub(html_pattern, '', x) if pd.notnull(x) else x)
```

```python
train_data.to_csv('cleaned_file.csv', index=False)
```

```python
columns_to_merge = ['TITLE', 'BULLET_POINTS', 'DESCRIPTION']

# Merge the columns into a single column
cleaned_data['merged_column'] = cleaned_data[columns_to_merge].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

```python

column7 = 'merged_column'
merged_column_null = cleaned_data[column7].isnull().sum()
print(f"Number of null values in '{column7}': {merged_column_null}")

```

```python

# Converting to the lower Case words

column7 = 'merged_column' 
cleaned_data[column7] = cleaned_data[column7].str.lower()
```
